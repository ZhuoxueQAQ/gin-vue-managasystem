## 添加到数据库

```python
import cpca
import json
import re
from decimal import Decimal
from uuid import uuid4


class DecimalEncoder(json.JSONEncoder):
    def default(self,o):
        if isinstance(o, Decimal):
            return float(o)
        super(DecimalEncoder,self).default(o)


def is_not_null(val):
    return val is not None

def formatRadio(val):
    try:
        return 0.0 if str(val) == 'nan' else float(val[:len(val)-1])
    except Exception as e:
        print(e, val)
        return 0.0
        

def formatAgency(val):
    return "无" if str(val) == 'nan' else val
def formatCurrency(val):
    if str(val) == 'nan':
        return Decimal(0)
    if not isinstance(val, str):
        return Decimal(val)
    format_sub = re.compile(r'[^\d.]+')
    try: 
        return Decimal(format_sub.sub('', val))
    except:
        return Decimal(0)
        
    

def formatDate(val):
    val = val[:-1] if val[-1] == '日' else val
    new_val = re.sub(r'[\u4e00-\u9fa5/]', "-", val).split("-")
    # 如果不是日期格式
    return "" if not len(new_val) == 3 else "-".join(new_val)

def init_project(projectGuid, clients, landingAgencies, partners) -> Project:
    return Project(projectGuid=projectGuid, projectName="111", projectManager="无", selectedArea="", chargeStandard="无", projectFund=0, categories="", createdDate="", projectNum="0", trainingMode="", trainingTimes="0", trainingNumOfPerson="0", trainingStartTime="", trainingEndTime="", contractNum="无", contractStartTime="", contractEndTime="", schoolManageAmount=0, devFundingAmount=0, welfareAmount=0, classPayAmount=0, outlayAmount=0, creator="", schoolManageRadio=0, devFundingRadio=0, welfareRadio=0, classPayRadio=0, paidAmount=0, unpaidAmount=0, clients=clients, landingAgencies=landingAgencies, partners=partners, payGin=0, payHin=0, payIin=0, payJin=0, payKin=0, payLin=0, payMin=0, payNin=0, payOin=0, payPin=0, payQin=0, payRin=0, paySin=0, payTin=0, payGout=0, payHout=0, payIout=0, payJout=0, payKout=0, payLout=0, payMout=0, payNout=0, payOout=0, payPout=0, payQout=0, payRout=0, paySout=0, payTout=0)

def init_incomestream(projectGuid, clients, landingAgencies, partners) -> IncomeStream:
    return IncomeStream(createdDate="", projectGuid=projectGuid, isBorrowed="", isOffset="", payUnit="", incomeAmount=0, invoice="", incomeValidCode="", clients=clients, landingAgencies=landingAgencies, partners=partners, invoiceIssueDate="", invoiceAmount=0, invoiceCount=0, invoiceCode="", splitAmountDate="", schoolManageAmount=0, devFundingAmount=0, welfareAmount=0, classPayAmount=0, payG=0, payH=0, payI=0, payJ=0, payK=0, payL=0, payM=0, payN=0, payO=0, payP=0, payQ=0, payR=0, payS=0, payT=0, remarks="")

def init_outlaystream(projectGuid, projectName, createDate, totalPay, remarks, clients, landingAgencies, partners) -> OutlayStream:
    return OutlayStream(createdDate=createDate, projectGuid=projectGuid, projectName=projectName, outlayProjectCode="", categories="", payG=0, payH=0, payI=0, payJ=0, payK=0, payL=0, payM=0, payN=0, payO=0, payP=0, payQ=0, payR=0, payS=0, payT=0,totalPay=totalPay, clients=clients, landingAgencies=landingAgencies, partners=partners, remarks=remarks)

def init_c_l_p(client_name, client_radio, client_amount, agency_name, agency_raddio, agency_amount, parter_name, parter_radio, parter_amount):
    clients = json.dumps([{
            "agencyName": formatAgency(client_name),
            "shareRadio": formatRadio(client_radio),
            "shareAmount": formatCurrency(client_amount)
        }], ensure_ascii=False, cls=DecimalEncoder)
    landingAgencies = json.dumps([{
            "agencyName": formatAgency(agency_name),
            "shareRadio": formatRadio(agency_raddio),
            "shareAmount": formatCurrency(agency_amount)
        }], ensure_ascii=False, cls=DecimalEncoder)
    partners = json.dumps([{
            "partnerName": formatAgency(parter_name),
            "shareRadio": formatRadio(parter_radio),
            "shareAmount": formatCurrency(parter_amount),
            "caculation":"比例",
            "partnerType":"无"
        }], ensure_ascii=False, cls=DecimalEncoder)
    return clients, landingAgencies, partners

def set_value(label_prop_dict, data_src, data_dst):
    for prop, label in label_prop_dict.items():
        now_val =  data_src[label]
        if not str(now_val) == 'nan':
            # 比例
            if prop[-5:] == "Radio":
                now_val = float(now_val[:len(now_val)-1])
            if prop[-6:] == "Amount" or prop == "projectFund":
                now_val = formatCurrency(now_val)
            if prop in ["createdDate", "contractStartTime", "contractEndTime", "splitAmountDate"]:
                now_val = formatDate(now_val)
            setattr(data_dst, prop, now_val)
    return data_dst

def add_data_to_db(project_dict, incomestream_dict, now_project_data, session):
    
    # 项目的委托方、落地机构和技术方
    project_cliets, project_landingAgencies, project_partners = init_c_l_p(
        now_project_data["委托方"],
        now_project_data["委托方分成比例"],
        "0",
        now_project_data["落地机构"],
        now_project_data["落地机构分成比例"],
        "0",
        now_project_data["技术方"],
        now_project_data["技术方分成比例"],
        "0"
    )
    
    # a project
    new_project = init_project(str(uuid4()), project_cliets, project_landingAgencies, project_partners)
    # incomestreams
    income_stream_df = pd.read_csv("./培训费导入平台收入流水.csv")
    # 赋值
    new_project = set_value(project_dict, now_project_data, new_project)

        
    # 查找对应的收入流水
    incomestreams = income_stream_df[new_project.projectName == income_stream_df["培训项目"]]
    # 有流水
    if not incomestreams.shape[0] == 0:
        # todo stream num
        # todo 名称和比例以最后一个流水的信息为准，金额以当前流水为准
        last_imcomestream = incomestreams.iloc[incomestreams.shape[0]-1]
        for _, now_stream in incomestreams.iterrows():
            # 初始化收入流水的委托方、落地机构、技术方
            income_clients, income_landingAgencies, icnome_partners =  init_c_l_p(
                last_imcomestream["委托方"],
                last_imcomestream["委托方分成比例"],
                now_stream["委托方分成金额"],
                last_imcomestream["落地机构"],
                last_imcomestream["落地机构分成比例"],
                now_stream["落地机构分成金额"],
                last_imcomestream["技术方"],
                last_imcomestream["技术方分成比例"],
                now_stream["技术方分成金额"]
            )
            # 初始化收入流水对象
            new_incomestream = init_incomestream(new_project.projectGuid, income_clients, income_landingAgencies, icnome_partners)

            # 对收入流水进行赋值
            new_incomestream = set_value(incomestream_dict, now_stream, new_incomestream)
            # todo 对每条输入流水，分别根据委托方、落地、技术方的返款时间和分成金额初始化支出流水，并在备注是委托方?还是什么的分成
            rebate_time_list = ["委托方分成", "落地机构分成", "技术方分成"]
            for index, rebate in enumerate(rebate_time_list):
                now_key = rebate + "返款时间"
                now_key_val = now_stream[now_key]
                # 有返款时间
                if not str(now_key_val) == 'nan' and not str(now_key_val)[0] == '无':
                    now_key_val = "2014/1/1" if now_key_val == "已返" else now_key_val
                    # id, 返款时间、分成金额、备注。初始化支出流水
                    # 支出流水的某个机构的分成（这里只可能是委托方、落地机构、技术方三选一)
                    outlay_clients, outlay_landingAgencies, outlay_partners = init_c_l_p(
                        last_imcomestream["委托方"],
                        last_imcomestream["委托方分成比例"],
                        now_stream["委托方分成金额"] if index == 0 else "0",
                        last_imcomestream["落地机构"],
                        last_imcomestream["落地机构分成比例"],
                        now_stream["落地机构分成金额"] if index == 1 else "0",
                        last_imcomestream["技术方"],
                        last_imcomestream["技术方分成比例"],
                        now_stream["技术方分成金额"] if index == 2 else "0",
                    )
                    new_outlaystream = init_outlaystream(new_project.projectGuid, new_project.projectName, formatDate(now_key_val), formatCurrency(now_stream[rebate + "金额"]), rebate + "返款", outlay_clients, outlay_landingAgencies, outlay_partners)
                    # 导入支出流水到数据库
                    session.add(new_outlaystream)
                    
            # 导入收入流水到数据库
            session.add(new_incomestream)
                    
        # 汇总流水的信息重新初始化项目的落地机构等等

        project_cliets, project_landingAgencies, project_partners = init_c_l_p(
            last_imcomestream["委托方"],
            last_imcomestream["委托方分成比例"],
            incomestreams["委托方分成金额"].apply(lambda c: formatCurrency(c)).sum(),
            last_imcomestream["落地机构"],
            last_imcomestream["落地机构分成比例"],
            incomestreams["落地机构分成金额"].apply(lambda c: formatCurrency(c)).sum(),
            last_imcomestream["技术方"],
            last_imcomestream["技术方分成比例"],
            incomestreams["技术方分成金额"].apply(lambda c: formatCurrency(c)).sum()
        )
        new_project.clients, new_project.landingAgencies, new_project.partners =  project_cliets, project_landingAgencies, project_partners
    
    # 累加培训项目的学校管理费、入账费用、未到账费用
    new_project.schoolManageAmount = incomestreams["学校管理费"].apply(lambda c:formatCurrency(c)).sum()
    new_project.paidAmount = incomestreams["收入金额"].apply(lambda c: formatCurrency(c)).sum()
    new_project.unpaidAmount = new_project.projectFund - new_project.paidAmount
        
    # 培训起止时间
    new_project.trainingStartTime, new_project.trainingEndTime = new_project.contractStartTime, new_project.contractEndTime
    # 获取所属地的abcode字符串茂名市电白区中小学教师信息技术应用能力提升工程2.0培训项目
    new_project.selectedArea = " ".join(list(filter(is_not_null, cpca.transform([now_project_data["省份"], now_project_data["市"], now_project_data["区"]])['adcode'])))
    
    
    # 导入项目到数据库
    session.add(new_project)
```
